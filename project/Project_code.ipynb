1. Horeca operating permits

```python
permits = pd.read_csv('/Users/victoria/Project-Week-2-Amsterdam/datasets/Horeca_operating_permits.csv')
```

```python
#clean it right away!
permits = permits.drop(['zaak_specificatie','o_tijden_terras_zo_do_van', 'o_tijden_terras_vr_za_tot', 'o_tijden_terras_vr_za_van', 'o_tijden_terras_zo_do_tot', 'openingstijden_vr_za_tot', 'openingstijden_vr_za_van', 'openingstijden_zo_do_tot', 'openingstijden_zo_do_van', 'status_tijdelijk_terras', 'tijdelijk_terras_details', 'publ_besluit_tijdelijk_terras', 'toestemming_tijdelijk_terras', 'terrasgeometrie'], axis=1)
permits
```

```python
#status_vergunning (status of license)
permits['status_vergunning'].unique()
#we have 'Granted', 'Not yet known', 'Requested', nan,'Partially granted'
```

```python
#we are interested only in 'Verleend' and 'Deels verleend' and the ones that will not expire soon
valid_permits = permits[(permits['status_vergunning'].isin(['Verleend','Deels verleend']))&
                        (permits['einddatum']>'2020-09-01')]
valid_permits                          
```

```python
#now let's see what types of places we have
valid_permits['zaak_categorie'].unique()
```

```python
#we stick to few of categories
filtered = valid_permits[(valid_permits['zaak_categorie'].isin(['Café', 'Restaurant', 'Alcoholvrij', 'Eethuis', 'Fastfood']))]
filtered
```

```python
#numbers by categories for the whole Amsterdam
filtered['zaak_categorie'].value_counts()
```
